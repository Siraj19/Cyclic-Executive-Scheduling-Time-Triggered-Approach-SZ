```python
import itertools

def generate_combinations(numbers, target_sum):
    result = []
    for r in range(1, len(numbers) + 1):
        combinations = itertools.combinations(numbers, r)
        for combination in combinations:
            if sum(combination) <= target_sum:
                result.append(combination)
    return result

# Example usage
numbers = [1, 2, 3, 4, 5]
target_sum = 8
combinations = generate_combinations(numbers, target_sum)
for combination in combinations:
    print(combination)
```

```python
import numpy as np

def get_matching_numbers(bool_array, number_array):
    matching_numbers = []
    for i, value in enumerate(bool_array):
        if value:
            matching_numbers.append(number_array[i])
    return np.array(matching_numbers)

# Example usage
bool_array = np.array([True, False, True, False, True])
number_array = np.array([1, 2, 3, 4, 5])

matching_numbers = get_matching_numbers(bool_array, number_array)
print(matching_numbers)
```
