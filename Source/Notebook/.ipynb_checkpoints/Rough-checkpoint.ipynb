```python
import itertools

def generate_combinations(numbers, target_sum):
    result = []
    for r in range(1, len(numbers) + 1):
        combinations = itertools.combinations(numbers, r)
        for combination in combinations:
            if sum(combination) <= target_sum:
                result.append(combination)
    return result

# Example usage
numbers = [1, 2, 3, 4.5, 5]
target_sum = 8
combinations = generate_combinations(numbers, target_sum)
for combination in combinations:
    print(combination)
```

```python
import itertools

def generate_combinations(numbers, target_sum, required_numbers):
    result = []
    max_combination = ()
    max_num_unique_elements = 0

    for r in range(1, len(numbers) + 1):
        combinations = itertools.combinations(numbers, r)
        for combination in combinations:
            if all(num in combination for num in required_numbers):
                current_sum = sum(combination)
                if current_sum <= target_sum:
                    num_unique_elements = len(set(combination))
                    if num_unique_elements > max_num_unique_elements:
                        max_num_unique_elements = num_unique_elements
                        max_combination = combination
                    elif num_unique_elements == max_num_unique_elements and len(combination) > len(max_combination):
                        max_combination = combination

    return max_combination if max_combination else "Cannot be done"

# Example usage
numbers =  np.array([5, 1, 2, 4, 2, 1, 3])
target_sum = 8
required_numbers = []
combination = generate_combinations(numbers, target_sum, required_numbers)
print(combination)
```

```python
import numpy as np

def get_matching_numbers(bool_array, number_array):
    matching_numbers = []
    for i, value in enumerate(bool_array):
        if value:
            matching_numbers.append(number_array[i])
    return np.array(matching_numbers)

# Example usage
bool_array = np.array([True, False, True, False, True])
number_array = np.array([1.5, 2, 3, 4, 5])

matching_numbers = get_matching_numbers(bool_array, number_array)
print(matching_numbers)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_gantt_chart(tasks, start_times, durations):
    fig, ax = plt.subplots()

    # Set y-axis limits and labels
    ax.set_ylim(-1, len(tasks))
    ax.set_yticks(range(len(tasks)))
    ax.set_yticklabels(tasks)

    # Generate a color map for the tasks
    colormap = plt.cm.get_cmap('tab20')
    num_colors = len(tasks)
    colors = colormap(np.linspace(0, 1, num_colors))

    # Plot the Gantt bars with different colors for each task
    for i, task in enumerate(tasks):
        ax.broken_barh([(start_times[i], durations[i])], (i-0.4, 0.8), facecolors=colors[i])

    # Determine the maximum end time
    max_end_time = max([start_time + duration for start_time, duration in zip(start_times, durations)])

    # Set x-axis limits and label
    ax.set_xlim(0, max_end_time)
    ax.set_xlabel('Time')

    # Set plot title
    ax.set_title('Gantt Chart')

    # Display the chart
    plt.show()

# Example data
tasks = ['Task 1', 'Task 2', 'Task 3']
start_times = [1, 4, 2]
durations = [3, 2, 4]

# Plot the Gantt chart with different colors for each task
plot_gantt_chart(tasks, start_times, durations)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_gantt_chart(tasks, start_times, durations, periods, offsets):
    fig, ax = plt.subplots()

    # Set y-axis limits and labels
    ax.set_ylim(-1, len(tasks))
    ax.set_yticks(range(len(tasks)))
    ax.set_yticklabels(tasks)

    # Generate a color map for the tasks
    colormap = plt.cm.get_cmap('tab20')
    num_colors = len(tasks)
    colors = colormap(np.linspace(0, 1, num_colors))

    # Plot the Gantt bars with different colors for each task
    for i, task in enumerate(tasks):
        period = periods[i]
        offset = offsets[i]
        bar_start = start_times[i]
        while bar_start <= max(start_times):
            ax.broken_barh([(bar_start, durations[i])], (i-0.4, 0.8), facecolors=colors[i])
            bar_start += period

    # Calculate the hyperperiod
    hyperperiod = np.lcm.reduce(periods)

    # Set x-axis limits and label
    ax.set_xlim(0, hyperperiod)
    ax.set_xlabel('Time')

    # Set plot title
    ax.set_title('Gantt Chart')

    # Display the chart
    plt.show()

# Example data
tasks = ['Task 1', 'Task 2', 'Task 3']
start_times = [1, 4, 2]
durations = [3, 2, 4]
periods = [5, 6, 7]
offsets = [0, 1, 2]

# Plot the Gantt chart for the hyperperiod duration
plot_gantt_chart(tasks, start_times, durations, periods, offsets)
```

```python
import math

def generate_unique_points(n):
    points = []
    offset = 2.0 / n  # Offset value to distribute points uniformly
    increment = math.pi * (3 - math.sqrt(5))  # Angle increment for spiral

    for i in range(n):
        y = i * offset - 1 + (offset / 2)  # Vertical position between 0 and 1
        radius = math.sqrt(1 - y * y)  # Radius of the circle at this y position

        # Calculate the golden angle
        theta = i * increment

        # Convert polar coordinates to Cartesian coordinates
        x = math.cos(theta) * radius
        z = math.sin(theta) * radius

        points.append((x, y, z))

    return points

# Example usage
num_points = 5
unique_points = generate_unique_points(num_points)
print(unique_points)
```
